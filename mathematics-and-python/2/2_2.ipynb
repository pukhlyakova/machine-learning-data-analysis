```python
import re
import numpy as np
import math
```

```python
def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)
```

```python
%matplotlib inline
from matplotlib import pylab as plt
```

```python
x = range(1, 16)
y = [f(v) for v in x]
```

```python
plt.plot(x, y)
plt.show()
```

Многочлен первой степени

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
res1 = np.linalg.solve(a, b)
print a
print b
print res1
```

```python
def f1(x, cof):
    return cof[0] + cof[1]*x

x = range(1, 16)
y = [f1(v, res1) for v in x]

plt.plot(x, y)
plt.show()
```

Многочлен второй степени

```python
a = np.array([[1, 1, 1], [1, 8, 8 **2 ], [1, 15, 15 ** 2]])
b = np.array([f(1), f(8), f(15)])
res2 = np.linalg.solve(a, b)
print a
print b
print res2
```

```python
def f2(x, cof):
    return cof[0] + cof[1]*x + cof[2]*x*x

x = range(1, 16)
y = [f2(v, res2) for v in x]

plt.plot(x, y)
plt.show()
```

Многочлен третьей степени

```python
a = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
b = np.array([f(1), f(4), f(10), f(15)])
res3 = np.linalg.solve(a, b)
print a
print b
print res3
```

```python
def f3(x, cof):
    return cof[0] + cof[1]*x + cof[2]*x*x + cof[3]*x*x*x

x = range(1, 16)
y = [f3(v, res3) for v in x]

plt.plot(x, y)
plt.show()
```
