```python
from math import sqrt 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm
%matplotlib inline
```

# Распределение Стьюдента

Сгенерируем выборку объемом 1000

```python
dist = t(2.74335149908)
sample = dist.rvs(1000)
```

Расчитаем теоретическую плотность распределения Стьюдента

```python
x = np.linspace(-10, 25,1000)
pdf = dist.pdf(x)
```

Постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины

```python
plt.plot(x, pdf, linewidth = 2.5)
plt.hist(sample, bins=50, normed=True)
plt.ylabel("Student's t-distribution")
plt.xlabel('$x$')
```

Для оценки распределение выборочного среднего необходимо узнать дисперсию и матожидание распределения Стьюдента.

```python
print "Матожидание: ", dist.mean()
print "Дисперсия: ", dist.var()
```

Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. 

```python
def generate(size):
    means = []
    for j in range(1000):
        rvs = dist.rvs(size)
        means.append(rvs.mean())
    
    norm_rv = norm(dist.mean(), sqrt(dist.var() / size))
    x = np.linspace(-10, 25, 1000)
    pdf = norm_rv.pdf(x)
    
    plt.plot(x, pdf)
    plt.hist(means, normed = True)
    plt.xlabel('$\\bar{X}_n$')
    plt.ylabel('$f(\\bar{X}_n)$')
    plt.title('n = %d' % size)
    plt.show()
```

```python
generate(5)
```

```python
generate(10)
```

```python
generate(50)
```

Легко заметить, что при росте n точность аппроксимации растет, что иллюстрирует работу ЦПТ
